```python
from typing import Annotated

from langchain_core.tools import tool
from langchain_core.tools.base import InjectedToolCallId
from langgraph.prebuilt import InjectedState
# Version: 0.3.5

def make_handoff_tool(*, agent_name: str):
    """Create a tool that can return handoff via a Command"""
    tool_name = f"transfer_to_{agent_name}"

    @tool(tool_name)
    def handoff_to_agent(
        state: Annotated[dict, InjectedState],
        tool_call_id: Annotated[str, InjectedToolCallId],
    ):
        """Ask another agent for help."""
        tool_message = {
            "role": "tool",
            "content": f"Successfully transferred to {agent_name}",
            "name": tool_name,
            "tool_call_id": tool_call_id,
        }
        return Command(
            # navigate to another agent node in the PARENT graph
            goto=agent_name,
            graph=Command.PARENT,
            # This is the state update that the agent `agent_name` will see when it is invoked.
            # We're passing agent's FULL internal message history AND adding a tool message to make sure
            # the resulting chat history is valid.
            update={"messages": state["messages"] + [tool_message]},
        )

    return handoff_to_agent
```

```python
from langgraph.graph import MessagesState, StateGraph, START, END
from langgraph.prebuilt import create_react_agent
from langgraph.types import Command

# model = ChatAnthropic(model="claude-3-5-sonnet-latest")
from langchain_openai import ChatOpenAI

model = ChatOpenAI()

# Define travel advisor ReAct agent
travel_advisor_tools = [
    # get_travel_recommendations,
    make_handoff_tool(agent_name="hotel_advisor"),
    make_handoff_tool(agent_name="flight_advisor"),
    # make_handoff_tool(agent_name="weather_advisor"),
    # make_handoff_tool(agent_name="food_advisor")
]

travel_advisor = create_react_agent(
    model,
    travel_advisor_tools,
    prompt=(
        """
        You are a travel expert. Your job is to:
1. Immediately recommend one specific destination based on the user’s request.
2. If the user request has nothing to do with travel advice, consider the following:
    - If the user question involves hotel recommendations, transfer to the 'hotel_advisor' for hotel suggestions.
    - If the user asks about flights from an origin city, transfer to the 'flight_advisor'.

You must always:
- Provide a clear recommendation in natural language
- Follow it with the appropriate tool call(s)
- Never ask the user questions like “Shall we proceed...?”, “Shall I proceed...”,  or “Would you like...?”
- Never wait for confirmation

Assume all instructions are actionable. Take initiative and always continue the flow.

        """
    ),
)


def call_travel_advisor(
    state: MessagesState,
) -> Command[Literal["hotel_advisor", "flight_advisor", "__end__"]]:
    # You can also add additional logic like changing the input to the agent / output from the agent, etc.
    # NOTE: we're invoking the ReAct agent with the full history of messages in the state
    return travel_advisor.invoke(state)
```

```python
# from langchain_community.tools.tavily_search import TavilySearchResults
# tavily_tool = TavilySearchResults()

# Define hotel advisor ReAct agent
hotel_advisor_tools = [
    # tavily_tool,
    # get_hotel_recommendations,
    make_handoff_tool(agent_name="travel_advisor"),
    make_handoff_tool(agent_name="flight_advisor"),
    # make_handoff_tool(agent_name="weather_advisor"),
    # make_handoff_tool(agent_name="food_advisor")
]

hotel_advisor = create_react_agent(
    model,
    hotel_advisor_tools,
    prompt=(
        "You are a helpful hotel expert. Given a destination, recommend 2–3 hotel options. "
        "Summarize each hotel option briefly in a numbered list. "
        "If the user has already mentioned or clearly needs flight options, call 'flight_advisor' immediately after presenting hotel results. "
        "If the destination is unclear or missing, call 'travel_advisor'. "
        "Do not ask the user to confirm anything — act immediately and always respond with a helpful message."

    ),
)

def call_hotel_advisor(
    state: MessagesState,
) -> Command[Literal["travel_advisor", "flight_advisor", "__end__"]]:
    return hotel_advisor.invoke(state)
```

```python
from langchain_community.tools.tavily_search import TavilySearchResults

tavily_tool = TavilySearchResults()

# Define flight advisor ReAct agent
flight_advisor_tools = [
    tavily_tool,
    # get_hotel_recommendations,
    make_handoff_tool(agent_name="travel_advisor"),
    make_handoff_tool(agent_name="hotel_advisor"),
    # make_handoff_tool(agent_name="weather_advisor"),
    # make_handoff_tool(agent_name="food_advisor")
]

flight_advisor = create_react_agent(
    model,
    flight_advisor_tools,
    prompt=(
        "You are a flight expert. "
    "Given an origin and destination, use the Tavily tool to search using a phrase like: "
    "`flights from San Francisco to Aruba` or `cheap flights to Cancun from NYC`. "
    "Then summarize 2 to 3 results in a numbered list with flight number, time, flight duration, and price. "
    "Use only what you find in the Tavily result. Do NOT make up any flight information. "
    "If the origin or destination is unclear, call 'travel_advisor'. "
    "If the user also asks for hotels, call 'hotel_advisor'. "
    "Always provide a clear, helpful message before calling another agent."

    ),
)

def call_flight_advisor(
    state: MessagesState,
) -> Command[Literal["travel_advisor", "hotel_advisor", "__end__"]]:
    return flight_advisor.invoke(state)
```

```python
builder = StateGraph(MessagesState)
builder.add_node("travel_advisor", call_travel_advisor)
builder.add_node("hotel_advisor", call_hotel_advisor)
builder.add_node("flight_advisor", call_flight_advisor)

# we'll always start with a general travel advisor
builder.add_edge(START, "travel_advisor")

graph = builder.compile()

from IPython.display import Image
display(Image(graph.get_graph().draw_mermaid_png()))
```

```python
from langchain_core.messages import convert_to_messages

def pretty_print_messages(update):
    if isinstance(update, tuple):
        ns, update = update
        # skip parent graph updates in the printouts
        if len(ns) == 0:
            return

        graph_id = ns[-1].split(":")[0]
        print(f"Update from subgraph {graph_id}:")
        print("\n")

    for node_name, node_update in update.items():
        print(f"Update from node {node_name}:")
        print("\n")

        for m in convert_to_messages(node_update["messages"]):
            m.pretty_print()
        print("\n")

```

```python
for chunk in graph.stream(
    {
        "messages": [
            [
                "user",
                """
                I want to go somewhere warm in the Caribbean. 
                Please search for some flights I can take to one of the places of your recommendation.
                """,
            ]
        ],
    },
    subgraphs=True,
):
    pretty_print_messages(chunk)
```

```python
for chunk in graph.stream(
    {
        "messages": [
            [
                "user",
                """
                I want to go somewhere warm in the Caribbean. 
                Where should I go?
                Also, do you have any hotel recommendations in that place and why?
                
                """,
                # Also, what are some nice hotels in that place that I should consider?
            ]
        ],
    },
    subgraphs=True,
):
    pretty_print_messages(chunk)
```

```python
for chunk in graph.stream(
    {
        "messages": [
            [
                "user",
                """
                If I want to fly from San Francisco to Cancun, what flights can i take?
                Also, any travel advice regarding traveling to Cancun? What makes it a good travel destination?

                """,
            ]
        ],
    },
    subgraphs=True,
):
    pretty_print_messages(chunk)
```
