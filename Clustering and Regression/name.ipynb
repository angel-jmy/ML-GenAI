# Load the data

```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
'''Load the data'''
data = pd.read_csv('data.csv')
data.head()
```

# Preprocessing

```python
'''Remove invalid data'''
data = data[data.UnitPrice > 0].copy()
data = data[data.Quantity > 0].copy()
data = data[data.Country != 'Unspecified'].copy()


'''Compute the total price'''
data['TotalPrice'] = data.Quantity * data.UnitPrice

'''Convert datetime'''
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format = '%m/%d/%Y %H:%M')
data['daysSince'] = (datetime.now() - data.InvoiceDate).dt.days


```

# Feature engineering

```python
users = data.groupby(by = 'CustomerID').agg(
    NumPurchases = ('Country', 'count'),
    TotalSpent = ('TotalPrice', 'sum'),
    AvgSpent = ('TotalPrice', 'mean'),
    FirstPurchase = ('InvoiceDate', 'min'),
    LastPurchase = ('InvoiceDate', 'max'),
    Country = ('Country', 'first')
).reset_index()

users.head()
```

```python
'''Adding features to the aggregated table'''
users['DaySince'] = (datetime.now() - users.LastPurchase).dt.days
users['Duration'] = (users.LastPurchase - users.FirstPurchase).dt.total_seconds() / 24 / 3600

users.head()
```
