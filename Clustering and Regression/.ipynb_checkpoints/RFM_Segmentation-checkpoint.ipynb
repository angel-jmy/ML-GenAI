## 2. Use RFM to make value group portrait for users? ML

**Unsupervised learning**

Conceptualize after clustering.

**EM vs K-Means**

### Related Concepts:
- **K**: Number of clusters  
- **Centroid**: The mean vector for each cluster

### The algorithm principle of k-means:
1. Select K points as the initial class center points.  
2. Assign each point to the nearest class center point, resulting in K classes, and then recalculate the center point for each class.  
3. Repeat step 2 until the class no longer changes, or the maximum number of iterations is set.

# Reorder clusters based on actual R value
def order_cluster(cluster_name, target_name, df, ascending=False):
    df_new = df.groupby(cluster_name)[target_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_name, ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_new = pd.merge(df, df_new[[cluster_name, 'index']], on=cluster_name)
    df_new = df_new.drop([cluster_name], axis=1)
    df_new = df_new.rename(columns={'index': cluster_name})
    return df_new

df_user = order_cluster('R-valued', 'R', df_user, False)

```python

# Convert user vector dictionary to DataFrame
user_df = pd.DataFrame.from_dict(user_vectors, orient='index')
user_df.columns = [f'emb_{i}' for i in range(user_df.shape[1])]
user_df['CustomerID'] = user_df.index

# Cluster users based on interest (can be replaced with HDBSCAN, etc.)
X = user_df[[col for col in user_df.columns if col.startswith('emb_')]]
kmeans = KMeans(n_clusters=5, random_state=42)
user_df['InterestCluster'] = kmeans.fit_predict(X)


```python
```python
# ==============================
# 3. RFM Calculation (Customer Value Segmentation)
# ==============================

NOW = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (NOW - x.max()).days,
    'InvoiceNo': 'nunique',
    'Amount': 'sum'
}).rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'Amount': 'Monetary'
}).reset_index()


```python
# RFM Scoring (1–5, higher is better)
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1]).astype(int)
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5]).astype(int)
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5]).astype(int)

rfm['RFM_Score'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']


```python
# Customer type classification (can be customized)
def rfm_level(df):
    if df['RFM_Score'] >= 13:
        return 'High-Value Customer'
    elif df['RFM_Score'] >= 9:
        return 'Potential Customer'
    elif df['RFM_Score'] >= 5:
        return 'Regular Customer'
    else:
        return 'Churn Risk'

rfm['UserType'] = rfm.apply(rfm_level, axis=1)


```python
# ==============================
# 4. Combine Interest Clusters + RFM Profiles
# ==============================

result = pd.merge(rfm, user_df[['CustomerID', 'InterestCluster']], on='CustomerID', how='left')

# Display result
print(result.head())
